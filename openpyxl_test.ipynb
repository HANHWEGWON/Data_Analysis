```python
from openpyxl import Workbook
# 엑셀을 위한 workbook
import requests
# 웹에서 요청받아오는 객체
from bs4 import BeautifulSoup
# 웹 사이트에서 원하는 데이터만 가져오기위한 객체

# 엑셀 만드는 역할을 합니다.
wb = Workbook()

# 엑셀에서 워크시트 생성
ws = wb.active
ws.title = 'Data'

# 데이터 추가
title_data = [''상품명', '판매가', 'url'']
# 엑셀 저장
ws.append(title_data)

# G마켓 검색창에 검색할 내용 input으로 받기.
x = input("검색어를 입력해주세요")
print("\n--------------------------")

#검색 페이지의 개수는 임의로 지정해주기.
for i in range(1, 3):
    print(f"{i}페이지 --------------------------------------")
    # x 는 검색키워드, i 는 검색 페이지의 페이지번호이다.
    url = f"https://browse.gmarket.co.kr/search?keyword={x}&p={i}"
    # url을 요청해 받아온후 응답 변수에 받는다.
    response = requests.get(url)
    text = response.text
    # 웹 파싱하기위한 객체 선언.
    soup = BeautifulSoup(text, 'html.parser')
    
    ## 파싱하는 코드블럭-------------------------------------
    list = soup.select('.box__information')
    
    for x in list:
        title = x.select_one('.text__item').text
        url = x.select_one('a.link__item')['href']
        money = x.select_one('strong').text
        data = [title, money, url]
        # ws.append(data)는 엑셀에 data 객체를 추가시키는 함수이다.
        ws.append(data)
    

# 마지막 엑셀 저장..!
wb.save('append_example.xlsx')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from time import sleep

chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get("https://shopping.naver.com/home")

```

```python

# 검색 아이콘을 클릭하여 검색 창을 확장합니다.
from openpyxl import Workbook

# 엑셀 만드는 역할을 합니다.
wb = Workbook()

# 엑셀에서 워크시트 생성
ws = wb.active
ws.title = 'Data'

title = ['제목', '가격', 'url']
ws.append(title)

search_icon = driver.find_element(By.CLASS_NAME, '_combineHeader_expansion_search_inner_1VxB3')
search_icon.click()

search_input = driver.find_element(By.CLASS_NAME, '_ac_input ')
search_input.send_keys("항공점퍼")
search_input.send_keys(Keys.RETURN)
sleep(3)

current_height = driver.execute_script("return window.scrollY")

while True:
    driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
    sleep(1)
    
    scrolled_height = driver.execute_script("return window.scrollY")
    
    if (current_height == scrolled_height):
        break
    
    current_height = scrolled_height

products = driver.find_elements(By.CSS_SELECTOR, '.product_list_item__b84TO')

for product in products:
    title = product.find_element(By.CSS_SELECTOR, '.product_info_tit__c5_pb').text
    print(f"제목 {title}")
    price = product.find_element(By.CSS_SELECTOR, '.product_num__iQsWh').text
    print(f"가격 {price}")
    # link = product.find_element(By.TAG_NAME, 'a').get_attribute('href')
    link = product.find_element(By.XPATH, '//a[contains(text(), "상품 바로가기")]').get_attribute('href')
    print(f"링크 {link}")
    ws.append([title, price, link])


wb.save('naver_shoping.xlsx')
```
