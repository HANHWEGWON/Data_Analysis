```python
def set_member_tel(user, phon_num, l):
    f = open("./member_tel.txt", "r")
    
    if user not in l:
        l.append(f"이름 = {user}\n전화번호 = {phon_num}\n------------------------")
    else :
        idx = l.index(user)
        l.remove(f"이름 = {user}\n전화번호 = {phon_num}\n------------------------")
        l.insert(idx, f"이름 = {user}\n전화번호 = {phon_num}\n------------------------")
    
    for sub in len(l):
        str +=sub

    f.write(str)
    f.close()
    

l = list()

while True:
    user = input("이름을 입력하세요.")
    pw = input("비밀번호를 입력하세요.")

    f = open("./member.txt", "r")

    find = f"이름 = {user}\n패스워드 = {pw}"
    str = f.read()

    if find in str:
        print("로그인 성공")
        phon_num = input("전화번호를 입력하세요.")
        set_member_tel(user, phon_num, l)
        break
    else :
        print("로그인 실패")

    f.close()
```
