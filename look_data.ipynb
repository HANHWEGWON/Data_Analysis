```python
import matplotlib.pyplot as plt


#plt.plot([1,2,3,4, 6], [2, 4, 5, 6, 10], 's')

font1 = {'family': 'fantasy', 'color':'deeppink', 'weight' : 'normal', 'size': 'xx-large'}
plt.plot([1,2,3,4], [1,4, 9, 16])

plt.xlabel('x', labelpad=30, loc='left', fontdict =font1)
plt.ylabel('y', loc='top')
plt.show()
```

```python
import numpy as np
x = np.arange(3)

plt.bar(x, [100, 400, 900], color=['r', 'g', 'b'], width=0.6)

plt.xticks(x, ['2018', '2019', '2020'])
plt.show()
```

```python
y= np.arange(3)

plt.barh(y, [100,400,900], color=['r', 'g', 'b'], height=0.6)
plt.yticks(y, ['2018', '2019', '2020'])
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, cmap="Spectral", alpha=0.5)
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
wedgeprops = { 'width' : 0.7, 'edgecolor':'k', 'linewidth':1}

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0.1], colors=['red', 'yellow', 'green', 'purple'], wedgeprops=wedgeprops)
```

```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.title('Monthly Sales Comparion (2019-2020)')
plt.plot(sales_2019, label="2019")
plt.plot(sales_2020, label="2020")
x = np.arange(12)
plt.xticks(x, months)
plt.legend(loc='best')
plt.xlabel('Month')
plt.ylabel('Sales')

```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

plt.title("Bar Chart")
plt.ylim(0,50)
x = np.arange(5)
plt.bar(x, data, width=0.6)
plt.xticks(x, categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
```

```python
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # Banana   

wedgeprops = { 'width' : 0.7, 'edgecolor':'k', 'linewidth':1}

plt.pie(sizes, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0], startangle=90, colors=['blue', 'yellow', 'green', 'red'], shadow=True)

```

```python
x = np.linspace(0, 10, 100)

y=np.sin(x)
z=np.cos(x)

plt.title('Sine and Cosine Functions')
plt.plot(x, y, color='red', label='sin(x)')
plt.plot(x, z, color='blue', label='cos(x)')
plt.grid(True)
plt.ylabel('Y')
plt.xlabel('X')

plt.legend()
```
